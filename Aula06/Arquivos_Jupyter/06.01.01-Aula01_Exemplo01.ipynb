Este √© um exemplo b√°sico de **clusteriza√ß√£o n√£o supervisionada** usando o algoritmo **KMeans**, com dados criados artificialmente.

### 1. Importa√ß√µes

- **`make_blobs`**: gera dados sint√©ticos (fict√≠cios) para testes.
- **`KMeans`**: algoritmo de clusteriza√ß√£o usado para agrupar os dados.
- **`matplotlib.pyplot`**: biblioteca para visualiza√ß√£o de gr√°ficos.

````python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
````

### 2. Gera√ß√£o dos Dados

- Cria **300 pontos** agrupados de duas dimens√µes em **4 centros** (clusters).
- `X` s√£o os dados (coordenadas x e y).
- `y` s√£o os r√≥tulos reais dos grupos (n√£o usados no aprendizado, pois √© n√£o supervisionado).

````python
X, y = make_blobs(n_samples=300, centers=6, random_state=0)

print (f"\n\n Dados das Coordenadas : \n\n{X[:5]}")
````

üéØ Finalidade da make_blobs()

- A fun√ß√£o `make_blobs()` do scikit-learn tem como finalidade gerar dados sint√©ticos (artificiais) com agrupamentos bem definidos, ou seja, pontos que se agrupam naturalmente em "bolhas" (blobs ). Esses dados s√£o muito √∫teis para testar e entender algoritmos de clusteriza√ß√£o , como o KMeans .

1. Criar dados de exemplo
    - √â ideal quando voc√™ ainda n√£o tem dados reais ou quer **testar rapidamente um modelo**.

2. Teste de algoritmos de clusteriza√ß√£o
    - Como os grupos s√£o claros e definidos, √© f√°cil visualizar e verificar se o algoritmo est√° funcionando corretamente.

3. Visualiza√ß√£o simples
    - Os dados gerados normalmente t√™m 2 ou 3 dimens√µes, permitindo **visualiza√ß√µes** f√°ceis com **gr√°ficos 2D ou 3D**.


üéØ Par√¢metros principais:

| Par√¢metro | O que faz |
|-----------|-----------|
| `n_samples` | N√∫mero total de pontos (registros) a serem gerados. Ex: 300. |
| `centers` | N√∫mero de grupos (clusters) desejados. Ex: 4. |
| `random_state` | Semente para o gerador aleat√≥rio, garantindo reproducibilidade. |
| `n_features` *(opcional)* | N√∫mero de caracter√≠sticas por ponto (padr√£o √© 2). |

üéØ Sa√≠das:
- **`X`**: matriz de features (coordenadas dos pontos).
- **`y`**: vetor com os r√≥tulos reais dos grupos (s√≥ usado para valida√ß√£o/teste ‚Äî n√£o √© necess√°rio no aprendizado n√£o supervisionado).

### 3. Configura√ß√£o e Treinamento do Modelo

‚úÖ O que √© esse c√≥digo?

- Este √© um exemplo de como usar o algoritmo **KMeans**, um m√©todo de **clusteriza√ß√£o n√£o supervisionada**, para **agrupar dados em 4 grupos**.

````python
kmeans = KMeans(n_clusters=6, random_state=0)
kmeans.fit(X)
````

O primeiro comando  **cria** o modelo `KMeans`.
- `n_clusters=4`: queremos dividir os dados em **4 grupos**.
- `random_state=42`: garante que os resultados sejam **sempre os mesmos** quando o c√≥digo for executado novamente (reprodutibilidade).

No segundo comando o modelo **aprende** a estrutura dos dados e **encontra os 4 grupos**.
- `X` √© uma matriz com os dados (ex: coordenadas x e y).
- O `.fit()` ajusta os centros dos clusters para melhor representar os grupos nos dados.

### 4. Previs√£o dos Grupos

- Usa o modelo treinado para **prever** qual grupo cada ponto pertence.

````python
labels = kmeans.predict(X)
````

‚úÖ O que ele faz?

- Este comando usa o modelo **KMeans j√° treinado** para **prever a qual grupo (cluster)** cada ponto do conjunto `X` pertence.


üîç Detalhando:

- **`kmeans`**: √© o modelo de clusteriza√ß√£o que voc√™ treinou anteriormente com `.fit(X)`.
- **`.predict(X)`**: pergunta ao modelo: *"Com base no que voc√™ aprendeu, qual √© o grupo de cada ponto em X?"*
- **`labels`**: √© uma vari√°vel que vai armazenar as **respostas** ‚Äî ou seja, qual **r√≥tulo de cluster** foi atribu√≠do a cada ponto.

üìå Exemplo pr√°tico:

- Suponha que `X` tenha 300 pontos. Ap√≥s rodar esse comando, `labels` ser√° um array com 300 valores (cada um sendo 0, 1, 2 ou 3), indicando a qual dos **4 grupos** cada ponto foi atribu√≠do.

### 5. Visualiza√ß√£o dos Resultados

‚úÖ O que ele faz?

- Este c√≥digo **cria um gr√°fico de dispers√£o (scatter plot)** para **visualizar os clusters** encontrados pelo algoritmo **KMeans**.

````python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("Clusteriza√ß√£o com KMeans")
plt.show()
````

- Mostra um gr√°fico com os pontos, **coloridos por grupo**.
- Cada cor representa um cluster encontrado pelo KMeans.


üîç Cria o gr√°fico de pontos
```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
```
- `X[:, 0]`: pega a **primeira coluna** dos dados (coordenada x de cada ponto).
- `X[:, 1]`: pega a **segunda coluna** dos dados (coordenada y de cada ponto).
- `c=labels`: define a cor de cada ponto com base no **grupo (cluster)** atribu√≠do pelo KMeans.
- `cmap='viridis'`: define a **paleta de cores** usada no gr√°fico (neste caso, uma escala de verde).
